```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
```

```python
models ={
    'Linear_reg' : LinearRegression(),
    'Decision Tree Regressor' : DecisionTreeRegressor(),
    'random Forest Regressor' : RandomForestRegressor(n_estimators = 100,random_state = 45),
    'Gradient Boost'          : GradientBoostingRegressor(n_estimators = 100 , random_state = 56),
    'Extrime Gredient Boost'  : XGBRegressor(n_estimators = 90,random_state = 50),
    'support vector machine'  : SVR()
}
```

```python
results = []

# Train and Evaluate Each Model
for name, model in models.items():
    model.fit(x_train, y_train)  # Train model
    y_pred = model.predict(x_test)  # Make predictions

    # Calculate Metrics
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    results.append([name, mse, mae, r2])
    print(f"{name}: RÂ² = {r2:.4f}, MSE = {mse:.4f}, MAE = {mae:.4f}")

```
